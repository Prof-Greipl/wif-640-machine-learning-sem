# A Single Neuron with Phython

```
import numpy as np 


def activation(input_, act_func):
    if act_func == 'relu':
        return np.maximum(input_, np.zeros(input_.shape))
    elif act_func == 'linear':
        return input_
    else:
        raise Exception('Activation function is not defined.')


class neuron:
  def __init__(self,input_dim, activation):
    self.activation = activation
    self.b = 0
    self.W = np.zeros( input_dim )
  
  def setWeight(self, W ):
    self.W = W
    
  def setBias(self, b ):
    self.b = b
    
  def getActivation(self, A ):
    Z =  np.add( np.dot(self.W, A), self.b)
    return activation(Z, self.activation)
  
  def print(self):
    print("Neuron Konfiguration--------")
    print(f"W = ")
    print(self.W)
    print(f"b = ")
    print(self.b)
    print("-----Finished Neuron Config.")
  

# Create the Neuron
myNeuron = neuron(2, 'relu')
myNeuron.setWeight([[3,-3]])
myNeuron.setBias([1])

myNeuron.print()

# Evaluate
aIn  = [[1], [1]]
aOut =myNeuron.getActivation( aIn )
print(aOut)

# This neuron works also for several inputs
aIn  = [[1,2], [2,3]]
aOut =myNeuron.getActivation( aIn )
print(aOut)



```
