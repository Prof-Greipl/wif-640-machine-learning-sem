# Load MPG Training set

`X` is a  $398 \times 7$ feature matrix, the features being stored in rows.
`y` returns  a $398 \times 1$ matrix, i.e. a row vector. It has one entry for each feature vector.

```
import pandas as pd
import numpy as np


def load_auto_mpg_data(samples = -1):
  df = pd.read_csv(
      "https://data.heatonresearch.com/data/t81-558/auto-mpg.csv", 
      na_values=['NA', '?'])


  print( df.describe() )

  # Handle the missing values
  df['horsepower'] = df['horsepower'].fillna(df['horsepower'].median())

  nCyl = df['cylinders']
  nDis = df['displacement']
  nHor = df['horsepower']
  nWei = df['weight']
  nAcc = df['acceleration']
  nYea = df['year']
  nOri = df['origin']

  X = pd.concat( [nCyl, nDis, nHor, nWei, nAcc, nYea, nOri], axis = 'columns').values;
  y_true = np.array( [ df['mpg'].values ] ) # regression

  if (samples < 0 ):
    return (X, y_true.T, df)
  else:
    return ( X[:, 0:samples], y_true[:, 0:samples].T, df)

(X, y_true, df) = load_auto_mpg_data()
print( X.shape )
print( y_true.shape )


```
