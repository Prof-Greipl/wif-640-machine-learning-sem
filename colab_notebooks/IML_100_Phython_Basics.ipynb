## Lists, Iterations

### Create a list

```
liste = ['A','Huber', True,1, 1.4]
print( liste )
```

### Appending elements.
Lists can have duplicates, elements are added at the end of a list

```
liste = ['A','Huber', True,1, 1.4]
liste.append('X')
liste.append('A')

print( liste )
```

### Inserting Elements

```
liste = ['A','Huber', True,1, 1.4]
liste.insert(0,'ZERO');
print( liste )
liste.insert(1,'NEU');
print( liste )
```

### Functions on lists

```
liste = ['A','Huber', True,1, 1.4]

# Length
print( len(liste) )

# Access a certain element
print( liste[3] )
```

### Iteration of Lists

```
liste = ['A','Huber', True,1, 1.4]
print( liste )
for s in liste:
  print(s)
  # print (s+1)    Explain, why this would fail
  # print (s+'x')  Explain, why this would fail
```
