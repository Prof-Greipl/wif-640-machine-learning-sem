# Loading Images (load_mnist())

```
# https://www.python-course.eu/neural_network_mnist.php
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def load_mnist():
  X = np.loadtxt("sample_data/mnist_test.csv", delimiter=",")
  print( X.shape  )
  return X
```

# Analyse Data

```
#Analyse the data
#Test
X = load_mnist()
display ( X[0:10])
```

# Display Image

```
def display_image ( row ):  
  img = row[1:].reshape(28,28)
  print( row[0] )
  plt.figure()
  plt.imshow(img, cmap="Greys")


for i in range(5):
  display_image( X[i] )
  
```

# Extract X and y

```
y = X[:,:1]
X = X[:,1:]
print (X.shape)
print (y.shape)
print (y[0:5])
```

# One-Hot-Encoding of y

```
import numpy as np
lr = np.arange(10)  # we have 10 labels
y_one_hot = (lr==y).astype(np.float)
y_one_hot[ y_one_hot == 1] = 1
y_one_hot[ y_one_hot == 0] = 0

print( y_one_hot.shape)

for i in range(5):
  print( y_one_hot[i])
```
