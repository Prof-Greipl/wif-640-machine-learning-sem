# Prep: Splitting Data with train_test_split

Note that splits are done one the first dimension, so lines of the matrix are splitted.

```
import numpy as np 
import pandas as pd
from sklearn.model_selection import train_test_split

X = np.arange(1,201).reshape(2,100)
y_true = np.arange(1,101).reshape(1,100) +500
print(y_true )

X_train, X_test, y_train, y_test = train_test_split(    
    X.T, y_true.T, test_size=0.05, random_state=42)

print(X_train.shape)
print(X_test.shape)
print(X_test)
print(y_test)
```
