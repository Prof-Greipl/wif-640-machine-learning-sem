# Neural Network Layer

```
import numpy as np 


def activation(z, act_func):
    global _activation
    if act_func == 'relu':
       return np.maximum(z, np.zeros(z.shape))
    
    elif act_func == 'sigmoid':
      return 1.0/(1.0 + np.exp( -z ))

    elif act_func == 'linear':
        return z
    else:
        raise Exception('Activation function is not defined.')


class layer:
  def __init__(self,input_dim, output_dim, activation):    
    self.activation = activation
    self.b = np.zeros( (output_dim,1) ) 
    self.W = np.zeros( (output_dim, input_dim) )
  
  def setWeight(self, W ):
    self.W = W
    
  def setBias(self, b ):
    self.b = b
    
  def getActivation(self, A ):
    Z =  np.add( np.dot(self.W, A), self.b)
    return activation(Z, self.activation)
  
  def print(self):
    print("Neuron Konfiguration--------")
    print(f"W = ")
    print(self.W)
    print(f"b = ")
    print(self.b)
    print("-----Finished Neuron Config.")
  
```

# Test

```
# Create a Layer
myLayer = layer(3,2,'relu')
myLayer.setWeight([[1,2], [2,-4]])
myLayer.setBias([[1],[1]])

myLayer.print()

# Evaluate
aIn  = [[1], [1]]
aOut =myLayer.getActivation( aIn )
print(aOut)

# This neuron works also for several inputs
aIn  = [[1,2], [2,3]]
aOut =myLayer.getActivation( aIn )
print(aOut)
```
