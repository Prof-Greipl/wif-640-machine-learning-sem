# Curves

```
# Plotting a function graph
# Visualozing Data see https://colab.research.google.com/notebooks/charts.ipynb#scrollTo=08RTGn_xE3MP
import numpy as np
import matplotlib.pyplot as plt

# Erzeuge Daten
x_data = np.linspace(-1,1, 100)
y_data = np.linspace(-1,1, 100)

plt.figure(figsize=(8,8))
plt.axis([-1.1, 1.1, -1.1, 1.1])
plt.axes().set_aspect(1)
plt.plot(x_data, x_data)
plt.plot(x_data, x_data * x_data)
plt.ylim([-1,1])  
plt.grid(True)
plt.show()
```

# Points

```
N = 1000
x_data = np.random.randn(N)
y_data = np.random.rand(N)

plt.figure()
plt.axes().set_aspect(1)
plt.axis([0, 1, 0, 1])
plt.scatter(x_data, y_data, marker = 'o') 
plt.show()
```

# Histograms

```
x = np.random.randn(10000000)
plt.hist(x, bins=1000)
plt.ylabel('No of times')
plt.show()
```
