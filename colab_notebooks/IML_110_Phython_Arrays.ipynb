# Features and Labels

Arrays are relevant, because we use $N \times d$ matrices to define feature vectors and $N \times L$ matrices to define the labels. In simple cases, we have $L=1$.

# Arrays and Shapes

```
import numpy as np

# One dimensional array (Vektor)
a = np.array([1,2,3,4])
print ("\n--- One dimensional array ---")
print (f'Werte      :{a}' )
print (f'Dimension  :{a.ndim}' )
print (f'Shape      :{a.shape}' )
print (f'Size       :{a.size}')
print (f'Datatype   :{a.dtype}')
print (f'Element [1] :{a[1]}')

# Two dimensional array (Matrix)
a = np.array([[1,2,3,4], [5,6,7,8]])
print ("\n--- Matrix ---")
print (f'Werte      :{a}' )
print (f'Dimension  :{a.ndim}' )
print (f'Shape      :{a.shape}' )
print (f'Size       :{a.size}')
print (f'Datatype   :{a.dtype}')
print (f'Element [1,2] :{a[1][2]}')

# Very simple two dimensional array
a = np.array([[2]])
print ("\n--- Weights (needed for simple Neuron ---")
print (f'Werte      :{a}' )
print (f'Dimension  :{a.ndim}' )
print (f'Shape      :{a.shape}' )
print (f'Size       :{a.size}')
print (f'Datatype   :{a.dtype}')
print (f'Element [0][0] :{a[0][0]}')

```

# Operations on Arrays

```
a = np.array([1,2,3])

print (a+1)
print (a * 10)
print (a/2)
print (a*a)
print (np.sqrt(a))
print (np.exp(a))
print (np.sin(a))
```

# Indexing, Slicing and Striding

### Access with Index

```
import numpy as np
a = np.arange(12);
print (a)
b = a.reshape(1,2,3);
print (b)
print( b[1] )
print( b[0][0])
```

## Slice and Stride

```
a = np.arange(10)
print (a[:5])
print (a[:-3])
print (a[-3:-2])


b = a.reshape(2,5)
print( b[1:2,:3] )

c = np.arange(10)
z = c[2:5]
z[0] = 100
print( z )
```
